```python
import matplotlib.pyplot as mp
import numpy as np
```

```python

x=[1,2,6,18]
y=[3,10,12,20]
mp.plot(x,y,color="green",marker="o")
mp.xlabel("X")
mp.ylabel("Y")
mp.title("Graph")
mp.xticks(np.arange(min(x),max(x)+1,2))#array passed
mp.yticks(np.arange(min(y),max(y)+1,2))
```

```python
# Write a Python program to create multiple plots. 
x = np.linspace(0, 10, 100) 
print(x)
y1 = np.sin(x)
y2 = np.cos(x)
mp.subplot(2, 1, 1) 
mp.plot(x, y1)
mp.title('Plot 1')
mp.subplot(2, 1, 2) 
mp.plot(x, y2)
mp.title('Plot 2')
mp.tight_layout() 
mp.show()
```

```python
#  Write a Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women. 
# Sample Data: 
# Means (men) = (22, 30, 35, 35, 26) 
# Means (women) = (25, 32, 30, 35, 29)

gps = ["G1","G2","G3","G4","G5"]
mean = (22, 30, 35, 35, 26)
means_w = (25, 32, 30, 35, 29)
x = np.arange(5)
bar_width = 0.3
mp.bar(x-bar_width/2,mean,bar_width,label="Men") 
mp.bar(x+bar_width/2,means_w,bar_width,label="Women")
mp.xticks(x, gps)
mp.xlabel("Groups")
mp.ylabel("Mean Height")
mp.legend() 
mp.show()
```
