```python
from sklearn import datasets
from sklearn import tree
from sklearn.model_selection import train_test_split


df=datasets.load_iris()
x=df.data
y=df.target




x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)


c=tree.DecisionTreeClassifier()



c.fit(x_train,y_train)
tree.plot_tree(c)

print(c.score(x_test,y_test))


feature = [[int(i) for i in input("sl,sw,pl,pw:").split()]]

r=c.predict(feature)
print(r)#reeturns index
print(df['target_names'][r])#a indexile target name
```

```python
from sklearn import datasets
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
df=datasets.load_iris()
x=df.data
y=df.target

x_test,x_train,y_test,y_train=train_test_split(x,y,train_size=0.7)

c=tree.DecisionTreeClassifier()
c.fit(x_train,y_train)
tree.plot_tree(c)

r=c.predict(x_test)
print(accuracy_score(r,y_test)*100)

f=[[int(i) for i in input("enter 4").split()]]
z=c.predict(f)
print(df['target_names'][z])
```
